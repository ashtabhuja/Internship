# Multicollinearity

We are checking the multicollinearity in our data

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
VIF=pd.DataFrame()
VIF['VIF Values']=[variance_inflation_factor(x.values,i)for i in range(len(x.columns))]
VIF['Features']=x.columns

VIF
```

# Modelling

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

Since we are using the regression method for oue machine learning hance we consider the R2 score

```python
maxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    pred=lr.predict(x_test)
    acc=r2_score(y_test,pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
        
print('The best accuracy is',maxAccu,'On random state',maxRS)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
print('Mean absolute error',mean_absolute_error(y_test,pred))
print('Mean squared error',mean_squared_error(y_test,pred))
print("Root mean square error",np.sqrt(mean_squared_error(y_test,pred)))
```

```python
print('R2 score',r2_score(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
rfpred=rf.predict(x_test)
print('Mean absolute error',mean_absolute_error(y_test,rfpred))
print('Mean sqaured error',mean_squared_error(y_test,rfpred))
print('Root mean squared error',np.sqrt(mean_squared_error(y_test,rfpred)))
print('R2 score',r2_score(y_test,rfpred))
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
dtpred=dt.predict(x_test)
print('Mean absolute error',mean_absolute_error(y_test,dtpred))
print('Mean squared error',mean_squared_error(y_test,dtpred))
print('Root mean square error',np.cbrt(mean_squared_error(y_test,dtpred)))
print('R2 sccore',r2_score(y_test,dtpred))
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor()
knn.fit(x_train,y_train)
knnpred=knn.predict(x_test)
print('Mean absolute erro',mean_absolute_error(y_test,knnpred))
print('Mean squared error',mean_squared_error(y_test,knnpred))
print('Root mean squre error',np.cbrt(mean_squared_error(y_test,knnpred)))
print('r2 score',r2_score(y_test,knnpred))
```

```python
from sklearn.ensemble import ExtraTreesRegressor
et=ExtraTreesRegressor()
et.fit(x_train,y_train)
etpred=et.predict(x_test)
print('Mean absolute error',mean_absolute_error(y_test,etpred))
print('Mean squared error',mean_squared_error(y_test,etpred))
print('Root mean squared erro',np.cbrt(mean_squared_error(y_test,etpred)))
print('r2 score',r2_score(y_test,etpred))
```
