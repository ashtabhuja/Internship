# 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
#Creating the blank list for fetching the data:
Brand_Name=[]
Product_Name=[]
Price=[]
Return=[]
Expected_delivery=[]
Availability=[]
Product_URL=[]
```

```python
#Getting the product URL:
Product_URL=[]
start=0
end=3
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        Product_URL.append(i.get_attribute('href'))
    nxt_button=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    nxt_button.click()
    time.sleep(2)
```

```python
len(Product_URL)
```

```python
for i in Product_URL:
    driver.get(i)
    time.sleep(3)
    try:
        brand_name=driver.find_element(By.XPATH,'//tr[@class="a-spacing-small po-brand"]/td[2]/span[1]')
        Brand_Name.append(brand_name.text)
    except NoSuchElementException:
        Brand_Name.append('-')
        
    
    
    try:
        product_name=driver.find_element(By.XPATH,'//span[@id="productTitle"]')
        Product_Name.append(product_name.text)
    except NoSuchElementException:
        Product_Name.append('-')
        
    
    
    try:
        price=driver.find_element(By.XPATH,'//span[@class="a-price aok-align-center reinventPricePriceToPayMargin priceToPay"]')
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('-')
    
    try:
        returns=driver.find_element(By.XPATH,'//li[3][@class="a-carousel-card tw-scroll-carousel-element"]')
        Return.append(returns.text)
    except NoSuchElementException:
        Return.append('-')
    
    
    try:
        availability=driver.find_element(By.XPATH,'//div[@class="a-section a-spacing-none a-spacing-top-micro }"]/span')
        Availability.append(availability.text)
    except NoSuchElementException:
        Availability.append('-')
        
    
    try:
        delivery=driver.find_element(By.XPATH,'//div[@id="mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE"]/span//span[1]')
        Expected_delivery.append(delivery.text)
    except NoSuchElementException:
        Expected_delivery.append('-')
        
    time.sleep(1)
```

```python
#length for the list:
print(len(Brand_Name),len(Product_Name),len(Price),len(Return),len(Availability),len(Expected_delivery))
```

```python
#Creating the data frame:

Products=pd.DataFrame({"Brand Name":Brand_Name,"Product Name":Product_Name,"Return":Return,"Expected Delivery":Expected_delivery,'Price':Price,'Availability':Availability})
Products
```

```python
#Creating the empty list:
Hostel_name=[]
Distance=[]
Ratings=[]
Total_reviews=[]
Overall_reviews=[]
Private_price=[]
Dorm_price=[]
Facilities=[]
Property_description=[]
product_url=[]
```

```python
#URL for the pages:
product_url=[]
url=driver.find_elements(By.XPATH,'//div[@class="property-card"]//a//a')
for i in url:
    href=i.get_attribute('href')
    product_url.append(href)
```

```python
for i in product_url:
    driver.get(i)
    time.sleep(5)
    
    try:
        hostel_name=driver.find_element(By.XPATH,'//div[@class="title-2"]')
        Hostel_name.append(hostel_name.text)
    except NoSuchElementException:
        Hostel_name.append('-')
        
    time.sleep(1)
    
    try:
        distance=driver.find_element(By.XPATH,'//div[@class="property-distance"]//span[2]')
        Distance.append(distance.text)
    except NoSuchElementException:
        Distance.append('-')
        
    time.sleep(1)
    
    try:
        rat=driver.find_element(By.XPATH,'//div[@class="flex-20"]/div[1]/div[1]')
        Ratings.append(rat.text)
    except NoSuchElementException:
        Ratings.append('-')
    time.sleep(1)
    
    try:
        total=driver.find_element(By.XPATH,'//div[@class="flex-20"]/div[1]/div[2]/div[2]')
        Total_reviews.append(total.text)
    except NoSuchElementException:
        Total_reviews.append('-')
    time.sleep(1)
    
    try:
        overall=driver.find_element(By.XPATH,'//div[@class="flex-20"]/div[1]/div[2]/div/span')
        Overall_reviews.append(overall.text)
    except NoSuchElementException:
        Overall_reviews.append('-')
    time.sleep(1)
    
    try:
        fac1 = driver.find_elements(By.XPATH,'//div[@id="facilities-section"]//div//ul//li[1]//ul[1]//li[2]')
        fac2 = driver.find_elements(By.XPATH,'//div[@id="facilities-section"]//div//ul//li[2]//ul[1]//li[17]')
        for i in fac1:
            for j in fac2:
                Facilities.append(i.text +', '+ j.text )
    except NoSuchElementException:
        Facilities.append('-')
    time.sleep(1)
    
    try:
        desc=driver.find_element(By.XPATH,'//div[@class="flex-80"]//div[4]')
        Property_description.append(desc.text)
    except NoSuchElementException:
        Property_description.append('-')
    time.sleep(1)
    
    try:
        private=driver.find_element(By.XPATH,'//div[5][@data-v-c308b52c]/div/div[1]/div[2]/strong[1]')
        if private:
            Private_price.append(private.text)
        else:
            Private_price.append('-')
    except NoSuchElementException:
        Private_price.append('-')
    time.sleep(1)
```
