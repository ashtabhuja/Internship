# Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.
This task will be done in following steps:
1. First get the webpage https://www.shine.com/
2. Enter “Data Analyst” in “Job title, Skills” field and enter “Bangalore” in “enter the location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

```python
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import time
```

Connecting to webdriver

```python
#connecting to the web driver
driver=webdriver.Chrome()
driver.get('https://www.shine.com/')
```

Sending input as Data analyst and Banglore in location

```python
#Enter the designation as Data Analyst
skills=driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[1]/div/input")
skills.send_keys("Data Analyst")
```

```python
#Enter the location as Banglore
location=driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
location.send_keys("Banglore")
```

```python
#Make empty list to scrap data
job_title=[] 
job_location=[] 
company_name=[]
experience_required=[]
```

Scarpping Data

```python
jobtags=driver.find_elements(By.XPATH,'//h2[@itemprop="name"]/a')
for i in jobtags:
    job=i.text
    job_title.append(job)
    
locations=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in locations:
    location=i.text
    job_location.append(location)
    
names=driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]/span')
for i in names:
    name=i.text
    company_name.append(name)
    
experiences=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in experiences:
    experience=i.text
    experience_required.append(experience)
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import time
```

```python
#Connecting to web driver:
driver10=webdriver.Chrome()
driver10.get('https://www.motor1.com/')
```

```python
#Entering the input as 50 most expensive cars and click on search using the xpath:

input10=driver10.find_element(By.XPATH,'/html/body/div[10]/div[2]/div/div/div[3]/div/div/div/form/input')
input10.send_keys('50 most expensive cars')

search10=driver10.find_element(By.XPATH,'/html/body/div[10]/div[2]/div/div/div[3]/div/div/div/form/button[1]')
search10.click()
```

```python
#Click on the 50 most expensive cars using the xpath:

click11=driver10.find_element(By.XPATH,'/html/body/div[10]/div[9]/div/div[1]/div/div/div[2]/div/div[1]/h3/a')
click11.click()
```

```python
#Scrapping data for Car name, Price for 50 cars:

Car_name=[]
Price=[]

car_name=driver10.find_elements(By.XPATH,'//h3[@class="subheader"]')[:50]
for i in car_name:
    cars=i.text
    Car_name.append(cars)
    
price=driver10.find_elements(By.XPATH,'//strong')
for i in price:
    prices=i.text
    Price.append(prices)
```

```python
print(len(Car_name),len(Price))
```

```python
#Creating data frame:

Car_list=pd.DataFrame({"Car Name":Car_name,"Price":Price})
Car_list
```
